```python
def norm_func(i):
    x = (i - i.min())	/ (i.max() - i.min())
    return (x)

df_norm = norm_func(df)
```

# Hierarchical clustering

```python
df_H=df_norm
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
z=linkage(df_norm, method='complete', metric='euclidean')
```

```python
plt.figure(figsize=(15,12));
dendrogram(z, leaf_rotation=0, leaf_font_size=10)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
h_complete= AgglomerativeClustering(n_clusters=3, linkage='complete', affinity='euclidean').fit(df_norm)
h_complete.labels_
```

```python
df_H['H_clusters']=pd.Series(h_complete.labels_)
```

```python
df_H.groupby(df_H.H_clusters).mean()
```

# KMeans

```python
df_K=df_norm
```

```python
from sklearn.cluster import KMeans
```

```python
TWSS=[]
k=list(range(2,9))
```

```python
for i in k:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_K)
    TWSS.append(kmeans.inertia_)
```

```python
plt.plot(k,TWSS,'ro-')
```

```python
model = KMeans(n_clusters = 3)
model.fit(df_K)
```

```python
model.labels_
```

```python
df_K['K_cluster']=pd.Series(model.labels_)
```

```python
df_K.groupby(df_K.K_cluster).mean()
```

# PCA

```python
df_PCA=df_norm
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 12)
pca_values = pca.fit_transform(df_PCA)
```

```python
var = pca.explained_variance_ratio_
var
```

```python
pca.components_
pca.components_[0]
```

```python
var1 = np.cumsum(np.round(var, decimals = 4) * 100)
var1
```

```python
plt.plot(var1, 'ro-')
```
